# データの読み込み

```python
!cd /home/masa1357/git/Kaggle-titanic
train_df = pd.read_csv('DATA/train.csv', index_col=0)
train_df
```

```python
test_df = pd.read_csv('DATA/test.csv', index_col=0)
test_df
```

# 前処理
前処理をするため、訓練データ（から目的変数を取り除いたもの）とテストデータを結合する

```python
df = pd.concat([train_df.drop(['Survived'], axis=1), test_df])
df
```

## 欠損処理
まず、欠損値の数を確認する

```python
df.isnull().sum()
```

`Cabin`は欠損率が高いので使用しない（補完しない）

`Age`と`Fare`は平均を補完する

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df
```

```python
df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
df
```

`Embarked`は最頻値を補完する

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df
```

もう一度欠損値を確認する

```python
df.isnull().sum()
```

## 特徴選択
使用する列だけを取り出す

`Name`と`Ticket`はユニーク性が高いので使用しない

```python
df = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
df
```

## ダミー変数化

```python
df = pd.get_dummies(df)
df
```

## 訓練データとテストデータに分割
訓練データの数を調べる

```python
nrow, ncol = train_df.shape
print(nrow, ncol)
```

先頭から訓練データの行数分だけデータを取り出して訓練データとし、目標変数の列を追加する

```python
train_df = df[:nrow].join(train_df['Survived'])
train_df
```

その次の行から最後までを取り出してテストデータとする

```python
test_df = df[nrow:]
test_df
```

# 機械学習
ここでは、とりあえず、決定木学習を使用する

```python
from sklearn.tree import DecisionTreeClassifier

X_train = train_df.drop(['Survived'], axis=1).to_numpy()
y_train = train_df['Survived'].to_numpy()

model = DecisionTreeClassifier(max_depth=3)
model.fit(X_train, y_train)
```

## 可視化
dtreevizを用いて学習した決定木を可視化する

```python
!pip3 -q install dtreeviz
```

```python
import dtreeviz

viz_model = dtreeviz.model(model, X_train=X_train, y_train=y_train,
    feature_names=train_df.columns,
    target_name='Survived', class_names=['Survived', 'Dead'])
viz_model.view()
```

# 予測
学習したモデルを用いて、テストデータに対して`Survived`の値を予測する

```python
X_test = test_df.to_numpy()

p_test = model.predict(X_test)
p_test
```

予測した値を提出用データに上書きする

```python
submit_df = pd.read_csv('/kaggle/input/titanic/gender_submission.csv', index_col=0)
submit_df['Survived'] = p_test
submit_df
```

CSVファイルに書き出す

```python
submit_df.to_csv('submission.csv')
```
